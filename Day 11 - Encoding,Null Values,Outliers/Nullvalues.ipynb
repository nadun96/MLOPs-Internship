# KNN Imputer

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=2)
titanic[["sibsp", "parch", "age"]] = imputer.fit_transform(
    titanic[["sibsp", "parch", "age"]]
)
```

# Outlier Ditection

```python
def outlier(data):
    threshold = 3
    mean= np.mean(data)
    sd = np.std(data)
    for i in range(len(data)):
        z_score = (int(data.iloc[i]- mean))/sd
        if np.abs(z_score) > threshold:
            # data = data.replace(i, mean)
            print(data.iloc[i] , z_score)
            # print("outliers found")
```

```python
# Z-score normalization
# Z = (X - mean) / std
# IQR normalization
# IQR = Q3 - Q1
outlier(titanic["fare"])
from sklearn.preprocessing import StandardScaler
```
