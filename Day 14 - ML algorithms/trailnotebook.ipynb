```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

mmscaler = MinMaxScaler()
sscaler = StandardScaler()
```

```python
data[["chol", "oldpeak", "thalach"]] = mmscaler.fit_transform(data[["chol", "oldpeak", "thalach"]])
```

```python
data.head()
```

```python
data[["age", "trestbps"]] = sscaler.fit_transform(data[["age", "trestbps"]])
```

```python
data.head()
```

```python
x = data.drop(["target"], axis=1)

y = data.target
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1 =  LogisticRegression()
```

```python
model1.fit(x_train, y_train)
```

```python
answer1 = model1.predict(x_test)
```

```python
from sklearn.svm import SVC

model2 = SVC()
```

```python
model2.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
score_lr = accuracy_score(y_test,answer1)
```

```python
print(score_lr*100)
```

```python
answer2 = model2.predict(x_test)
```

```python
svm_ac = accuracy_score(y_test,answer2)
print(svm_ac*100)
```
