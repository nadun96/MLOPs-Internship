```python
def outlier_handler(col):
    d = x[col]
    q1 = d.quantile(0.25)
    q3 = d.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    
    d = d.clip(lower, upper)
```
