```python
words = nltk.word_tokenize(sample)
print(words)
```

```python
print(len(words))
```

```python
from nltk.corpus import stopwords
stop = set(stopwords.words("english"))
```

```python
print(stop)
```

```python
new = []

for y in words:
    if y not in stop:
        new.append(y)
```

```python
new
```

```python
" ".join(new)
```

```python
import string
punctuations = list(string.punctuation)
```

```python
new = []

for y in words:
    if y not in (stop and punctuations):
        new.append(y)
```

```python
" ".join(new)
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
new2 = []

for n in new:
    stemmed_word = stemmer.stem(n)
    new2.append(stemmed_word)
```

```python
new2
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
new3 = []
for n in new:
    lem_word = lemmatizer.lemmatize(n)
    new3.append(lem_word)
```

```python
new3
```

```python
new_sentence = " ".join(new3)
```

```python
new_sentence
```

```python
new_sentence = nltk.sent_tokenize(new_sentence)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer()
```

```python
numformat = cv.fit_transform(new_sentence).toarray()
```
