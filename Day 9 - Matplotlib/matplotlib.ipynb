```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0,50])
y = np.array([0,50])
plt.plot(x, y)
plt.show()
```

```python
x = np.array([2,6,3,10])
plt.plot(x, linestyle='dotted')
plt.show()
```

```python
y = np.array([1,3,5,7])
x = np.array([2,6,3,10])
plt.plot(x, y, marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph')
plt.grid()
plt.show()
```

```python
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
plt.subplot(1, 2, 1)
plt.plot(x,y)

x = np.array([0,1,2,3])
y = np.array([10,20,30,40])
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.show()
```

```python
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
plt.subplot(2, 1, 1) # 2 rows, 1 column, 1st plot
plt.plot(x,y)

x = np.array([0,1,2,3])
y = np.array([10,20,30,40])
plt.subplot(2, 1, 2)
plt.plot(x,y)
plt.show()
```

```python
import pandas as pd 

df = pd.read_csv('heart/heart.csv')
```

```python
df.plot()
```

```python
df.head()
```

```python
plt.scatter(df['age'], df['target'])
plt.show()
```

```python
df.target == 0 # filter rows where target is 0
```

```python
df[df.target == 0] # filter rows where target is 0
```

```python
df[df.target == 0].age # filter rows where target is 0 and get age column
df[df.target == 1].age.hist() # histogram of age column where target is 1

plt.scatter(x = df[df.target == 0].age,y = df[df.target == 0].chol, color='red', label='Absent')
plt.scatter(x = df[df.target == 1].age, y = df[df.target == 1].chol, color='blue', label='Present')
plt.legend()
plt.title('Heart Disease age vs cholestrol')
plt.ylabel('Cholestrol')
plt.xlabel('Age')

plt.show()
```

```python
df[df.target == 0].age.hist(alpha=0.5, color='red', bins=30)
```

```python
plt.barh(df[df.target == 0].age, df[df.target == 0].chol, color='red', label='Absent')
```

```python
y = np.random.normal(0, 1, 1000)
plt.hist(y, bins=30)
```

```python
df.age.plot(kind="box")
```

```python
plt.boxplot([df[df.target == 0].age, df[df.target == 1].age], labels=['Absent', 'Present'])
plt.show()
```

```python
plt.pie([df[df.target == 0].age.count(), df[df.target == 1].age.count()], labels=['Absent', 'Present'], autopct='%1.1f%%', colors=['green', 'blue'])
plt.title('Heart Disease')
plt.show()
```
